## Two Characters
Given a string, remove characters until the string is made up of any two alternating characters. 
When you choose a character to remove, all instances of that character must be removed. 
Determine the longest string possible that contains just two alternating letters.

```python
from collections import Counter
from itertools import combinations

def alternate(s):
    # get all characters
    chars_list = list(s)
    chars = list(set(chars_list))

    # get frequency of every pair
    max_frequency =  0
    for char1, char2 in combinations(chars, 2):
        temp_list = [c for c in s if c in (char1, char2)]

        # check if elements are alternating
        def is_alternating(lst):
            for i in range(len(lst) - 1):
                if lst[i] == lst[i + 1]:
                    return False
            return True

        if is_alternating(temp_list):
            if len(temp_list) > max_frequency:
                max_frequency = len(temp_list)
    return max_frequency
    
alternate('abaacdabd')
```
