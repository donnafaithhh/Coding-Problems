## Rolling Dice

```python
# roll dice
die1 = random.randint(1, 6)
die2 = random.randint(1, 6)
print(f"die 1: {die1}\ndie 2: {die2}")
```

```python
counter_seven = 0
total_runs = 1_000_000

for i in range(total_runs):
    # roll dice
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    
    # check lucky 7
    total = die1 + die2
    if total == 7:
        counter_seven += 1

print(f"Probability of a lucky 7: {counter_seven / total_runs}")
```
