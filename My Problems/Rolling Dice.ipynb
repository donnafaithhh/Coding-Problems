## Rolling Dice

```python
# roll dice
die1 = random.randint(1, 6)
die2 = random.randint(1, 6)
print(f"die 1: {die1}\ndie 2: {die2}")
```

```python
counter_seven = 0
total_runs = 10_000_000

for i in range(total_runs):
    # roll dice
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    
    # check lucky 7
    total = die1 + die2
    if total == 7:
        counter_seven += 1

print(f"Probability of a lucky seven: {counter_seven / total_runs}")
```

```python
counter_duplicates = 0
total_runs = 10_000_000

for i in range(total_runs):
    # roll dice
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    die3 = random.randint(1, 6)
    
    # check if there are at least 2 same
    if (
        die1 == die2 or
        die1 == die3 or 
        die2 == die3
    ):
        counter_duplicates += 1

print(f"Probability of at least 2 pairs: {counter_duplicates / total_runs}")
```

```python
counter_same = 0
total_runs = 10_000_000

for i in range(total_runs):
    # roll dice
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    die3 = random.randint(1, 6)
    
    # check if all are same
    if die1 == die2 == die3:
        counter_same += 1

print(f"Probability of all being the same: {counter_same / total_runs}")
```
