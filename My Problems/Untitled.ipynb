```python
def count_bits_and_positions(n):
    """
    Returns an array where:
    - First element is the count of 1-bits in n's binary representation
    - Remaining elements are positions of 1-bits from left to right (starting at 1)
    
    Example: n = 37 → binary '100101' → returns [3, 1, 4, 6]
    """
    # Convert to binary string without '0b' prefix
    binary_str = bin(n)[2:]
    
    # Count 1-bits
    count = binary_str.count('1')
    
    # Find positions of 1-bits (leftmost is position 1)
    positions = [i + 1 for i, bit in enumerate(binary_str) if bit == '1']
    
    # Return as required format: [count, positions...]
    return [count] + positions

count_bits_and_positions(37)
```

```python
def list_to_roman(nums):
    """
    Convert a list of integers to a list of Roman numeral strings.
    
    Args:
        nums: List of integers (positive, typically 1-3999)
    
    Returns:
        List of Roman numeral strings
    """
    # Dictionary of value to symbol mappings (largest to smallest)
    roman_map = {
        1000: 'M',
        900: 'CM',
        500: 'D',
        400: 'CD',
        100: 'C',
        90: 'XC',
        50: 'L',
        40: 'XL',
        10: 'X',
        9: 'IX',
        5: 'V',
        4: 'IV',
        1: 'I'
    }
    
    # Helper function to convert a single number
    def int_to_roman(num):
        """Convert a single integer to Roman numeral using the dictionary."""
        if not isinstance(num, int) or num <= 0:
            return "Invalid"
        
        result = ""
        # Process values in descending order
        for value, symbol in sorted(roman_map.items(), reverse=True):
            while num >= value:
                result += symbol
                num -= value
        return result
    
    # Convert each number in the list
    return [int_to_roman(n) for n in nums]

list_to_roman([5, 1, 2, 3, 4, 5])
```
