## Cards

```python
# pick a card with replacement
suits = ['Heart', 'Spade', 'Diamond', 'Clover']
number = ['Ace', 2, 3, 4, 5, 6, 7, 8, 9, 10, 'Jack', 'Queen', 'King']
suit1, number1 = random.choice(suits), random.choice(number)

print(suit1, number1)
```

```python
# pick a card without replacement
suits = ['Heart', 'Spade', 'Diamond', 'Clover']
numbers = ['Ace', 2, 3, 4, 5, 6, 7, 8, 9, 10, 'Jack', 'Queen', 'King']

# create deck
deck = []
for suit in suits:
    for number in numbers:
        deck.append((suit, number))

# remove the drawn ones
drawn_pairs = []

card = random.choice(deck)
deck.remove(card)
drawn_pairs.append(card)
```
