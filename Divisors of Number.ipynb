## Find All Unique Divisors of a Number

Write a function get_divisors(n) that takes a positive integer n and returns a sorted list of all unique divisors of n (excluding 0).

Examples:

Input: 12
Output: [1, 2, 3, 4, 6, 12]

```python
def get_divisors(n):
    divisors = []
    for num in range(1, n+1):
        if n % num == 0:
            divisors.append(num)
    return divisors
```

```python
get_divisors(100)
```
