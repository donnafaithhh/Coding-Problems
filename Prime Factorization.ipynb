## Prime Factorization

Write a function a positive integer n and returns a sorted list of all prime factors.

```python
# find prime numbers with Wilson's Theorem
def is_prime(n):
    nums = list(range(1, n))
    factor = np.prod(nums)
    return (factor + 1) % n == 0

# for prime factorization
def prime_factor(n):
    factors = []
    prime_factors = []
    for i in range(1, n+1):
        if is_prime(i) and i > 1:
            prime_factors.append(i)

    for i in prime_factors:
        if n % i == 0:
            factors.append(i)
    
    return factors
```

```python
# finding prime numbers
def is_prime(n):
    nums = np.arange(2, n)
    for num in nums:
        if n % num == 0:
            return False
    return True
    
# for prime factorization
def prime_factor(n):
    if n < 2:
        return []
    
    factors = []
    temp = n
    i = 2
    while i * i <= temp:
        if is_prime(i):
            while temp % i == 0:
                factors.append(i)
                temp //= i
        i += 1

    if temp > 1:
        factors.append(temp)
    return factors
```
