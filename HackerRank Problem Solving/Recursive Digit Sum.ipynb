## Recursive Digit Sum
Get the super digit of a given number.

```python
def superDigit(n, k):
    # get individual digits
    def get_digits(n):
        return [int(digit) for digit in str(n)]
    
    digits = get_digits(n)
    k_digits = digits * k 
    while len(k_digits) > 1:
        temp_sum = sum(k_digits)
        k_digits = get_digits(temp_sum)
        print(temp_sum)
    return k_digits[0]
    
superDigit(9875, 4)
```

```python
def superDigit(n, k):
    # Initial digit sum multiplied by k
    n = str(n)
    total = sum(int(d) for d in n) * k

    # Reduce to super digit
    while total >= 10:
        total = sum(int(d) for d in str(total))

    return total
superDigit(9875, 4)
```
