## Re.split()

Sample input: 100,000,000.000

Sample output: \
100 \
000 \
000 \
000 

```python
pattern = r'\d+'
user_input = input()
all_digits = re.findall(pattern, user_input)
for group in all_digits:
    print(group)
```

```python
regex_pattern = r"[.,]"	# Do not delete 'r'.

import re
user_input = '100,000,000.000'
print("\n".join(re.split(regex_pattern, user_input)))
```
