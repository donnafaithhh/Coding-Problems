```python
import re
```

## Replacing parts of strings

```python
pattern = r"(\d\d\d)-(\d\d\d)-(\d\d\d\d)"
new_pattern = r"\1\2\3"
user_input = input()
print(re.sub(pattern, new_pattern, user_input))
```

## Verifying emails

```python
pattern = r"[a-zA-Z0-9]+@[a-zA-Z]+\.(com|edu|net)"
user_input = input()
bool(re.search(pattern, user_input))
```
