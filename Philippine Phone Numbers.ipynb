## Reverse String

Write a function that accepts a string, and returns the string in reverse format.

```python
# function that reverses a string
def reverse_string(string):
    return string[::-1]

reverse_string('ABC123')
```

## Reverse Words

Write a function that accepts a sentence, and returns the sentence in reverse format.

```python
# function that reverses a sentence
def reverse_sentence(sentence):
    splt_sentence = sentence.strip().split(' ')
    return splt_sentence[::-1]

reverse_sentence('A quick brown fox jumps over the lazy dog')
```
