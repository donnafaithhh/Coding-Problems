## Beautiful Pairs


```python
from itertools import combinations 

def beautifulPairs(A, B):
    # get beautiful set
    beautiful_set = []
    for idx, element1 in enumerate(A):
        for jdx, element2 in enumerate(B):
            if element1 == element2:
                beautiful_set.append((idx, jdx))
                
    return beautiful_set

A = [10, 11, 12, 5, 14]
B = [8, 9, 11, 11, 5]
beautifulPairs(A, B)
```
